```python
ind_list=['candle_1', 'candle_2', 'candle_3', 'anomaly', 'level']
def get_ind_data(symbol,start_date=datetime.datetime(2017,1,1).strftime("%Y-%m-%d")):
    df_all_ind=pd.DataFrame()
    for ind in ind_list:
#         print(ind)
        sql="select d.date_price as date, d.value from d_data d join symbol s on d.symbol_id = s.id join indicator i on i.id=d.indicator_id where s.ticker='%s' and i.name = '%s' and d.date_price > '%s'" %(symbol, ind, start_date)
        df_indicator=pd.read_sql(sql,con=conn_indicator)
        if df_indicator.empty:
            continue
        df_indicator.set_index('date',inplace=True)
        df_indicator=pd.concat([df_indicator.drop(['value'], axis=1), df_indicator['value'].apply(pd.Series)], axis=1)
        df_indicator.columns=[ind+"_"+col for col in df_indicator.columns]
        if df_all_ind.empty:
            df_all_ind=df_indicator
        else:
            df_all_ind=pd.merge(left=df_all_ind, right=df_indicator,on='date')

    df_all_ind.reset_index(inplace=True)
    df_all_ind.rename(columns={'anomaly_close':'close','anomaly_low':'low','anomaly_high':'high','anomaly_open':'open','anomaly_volume':'volume'},inplace=True)
    return df_all_ind
```

```python
%%time
interested_symbols=['AAN','ABCB']

win_dict={'num':[],'bars':[],'vol_ano':[],'gap_ano':[]}
loss_dict={'num':[],'bars':[],'vol_ano':[],'gap_ano':[]}
complete_dict={'vol_ano':[],'gap_ano':[]}
tp_mult=2

for symbol in interested_symbols:
    df_all_ind=get_ind_data(symbol=symbol)
    total_length=len(df_all_ind)
    print("Working On : ",symbol)
    complete_dict['gap_ano'].extend(list(df_all_ind['anomaly_gap_signal']))
    complete_dict['vol_ano'].extend(list(df_all_ind['anomaly_vol_anomaly']))
    for i, row in df_all_ind.iterrows():
        if row['candle_2_pattern_name']=='BULLISH ENGULFING':
            sl=row['candle_2_low']-row['anomaly_atr']/2
            tp=row['close']+tp_mult*(row['close']-sl)
            j=i
            while j <total_length:
                if df_all_ind.iloc[j]['low']<=sl:
                    loss_dict['num'].append(1)
                    loss_dict['bars'].append(j-i)
                    loss_dict['vol_ano'].append(df_all_ind.iloc[j]['anomaly_vol_anomaly'])
                    loss_dict['gap_ano'].append(df_all_ind.iloc[j]['anomaly_gap_signal'])
                    break
                elif df_all_ind.iloc[j]['high']>=tp:
                    win_dict['num'].append(1)
                    win_dict['bars'].append(j-i)
                    win_dict['vol_ano'].append(df_all_ind.iloc[j]['anomaly_vol_anomaly'])
                    win_dict['gap_ano'].append(df_all_ind.iloc[j]['anomaly_gap_signal'])
                    break
                else:
                    j+=1
        elif row['candle_2_pattern_name']=='BEARISH ENGULFING':
            sl=row['candle_2_high']+row['anomaly_atr']/2
            tp=row['close']-tp_mult*(sl-row['close'])
            j=i
            while j <total_length:
                if df_all_ind.iloc[j]['high']>=sl:
                    loss_dict['num'].append(1)
                    loss_dict['bars'].append(j-i)
                    loss_dict['vol_ano'].append(df_all_ind.iloc[j]['anomaly_vol_anomaly'])
                    loss_dict['gap_ano'].append(df_all_ind.iloc[j]['anomaly_gap_signal'])
                    break
                elif df_all_ind.iloc[j]['low']<=tp:
                    win_dict['num'].append(1)
                    win_dict['bars'].append(j-i)
                    win_dict['vol_ano'].append(df_all_ind.iloc[j]['anomaly_vol_anomaly'])
                    win_dict['gap_ano'].append(df_all_ind.iloc[j]['anomaly_gap_signal'])
                    break
                else:
                    j+=1
```
