```python
ind_list=['anomaly']
def get_ind_data(symbol,start_date=datetime.datetime(2017,1,1).strftime("%Y-%m-%d")):
    df_all_ind=pd.DataFrame()
    for ind in ind_list:
#         print(ind)
        sql="select s.ticker,d.date_price as date, d.value from d_data d join symbol s on d.symbol_id = s.id join indicator i on i.id=d.indicator_id where s.ticker='%s' and i.name = '%s' and d.date_price > '%s'" %(symbol, ind, start_date)
        df_indicator=pd.read_sql(sql,con=conn_indicator)
        if df_indicator.empty:
            continue
        df_indicator.set_index('date',inplace=True)
        df_indicator=pd.concat([df_indicator.drop(['value'], axis=1), df_indicator['value'].apply(pd.Series)], axis=1)
        df_indicator.columns=[ind+"_"+col for col in df_indicator.columns]
        if df_all_ind.empty:
            df_all_ind=df_indicator
        else:
            df_all_ind=pd.merge(left=df_all_ind, right=df_indicator,on='date')

    df_all_ind.reset_index(inplace=True)
    df_all_ind.rename(columns={'anomaly_close':'close','anomaly_low':'low','anomaly_high':'high','anomaly_open':'open','anomaly_volume':'volume'},inplace=True)
    return df_all_ind
```

```python
# df_symbol.iloc[i,df_symbol.columns.get_loc('target')]
```

```python
%%time
interested_symbols=['AAN','ABCB']
tp_mult=2

df_all_symbol=pd.DataFrame()
for symbol in interested_symbols:
    df_symbol=get_ind_data(symbol=symbol)
    total_length=len(df_symbol)
    df_symbol['target']=None
    print("Working On : ",symbol)
    df_symbol['open_r']=df_symbol['open']/df_symbol['open'].shift(1)-1
    df_symbol['close_r']=df_symbol['close']/df_symbol['close'].shift(1)-1
    df_symbol['high_r']=df_symbol['high']/df_symbol['high'].shift(1)-1
    df_symbol['low_r']=df_symbol['low']/df_symbol['low'].shift(1)-1
    df_symbol['gap_r']=df_symbol['open']/df_symbol['close'].shift(1)-1 # this would capture the gap
    df_symbol['body_r']=df_symbol['close']/df_symbol['open']-1 # this would capture the body size
    df_symbol['top_wick_r']=df_symbol['anomaly_wick_bull']/df_symbol['high'] # would capture the upper and lower wick kind of normalized from the high and low
    df_symbol['bottom_wick_r']=df_symbol['anomaly_wick_bear']/df_symbol['low']
    df_symbol['volume_r']=df_symbol['volume']/df_symbol['volume'].shift(1)-1
    for i, row in df_symbol.iterrows():
        sl=row['low']-row['anomaly_atr']/2
        tp=row['close']+tp_mult*(row['close']-sl)
        j=i
        while j <total_length:
            if df_symbol.iloc[j]['low']<=sl:
                df_symbol.iloc[i,df_symbol.columns.get_loc('target')]=-1
                break
            elif df_symbol.iloc[j]['high']>=tp:
                df_symbol.iloc[i,df_symbol.columns.get_loc('target')]=1                    
                break
            else:
                j+=1
        sl=row['high']+row['anomaly_atr']/2
        tp=row['close']-tp_mult*(sl-row['close'])
        j=i
        while j <total_length:
            if df_symbol.iloc[j]['high']>=sl:
                df_symbol.iloc[i,df_symbol.columns.get_loc('target')]=-1
                break
            elif df_symbol.iloc[j]['low']<=tp:
                df_symbol.iloc[i,df_symbol.columns.get_loc('target')]=1
                break
            else:
                j+=1
#     print(df_symbol.columns)
    if df_all_symbol.empty:
        df_all_symbol=df_symbol
    else:
        df_all_symbol=pd.concat([df_symbol,df_all_symbol])
```
