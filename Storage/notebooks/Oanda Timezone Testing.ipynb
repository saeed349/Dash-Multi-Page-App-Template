```python


# The v20 api handles from times a little differently - be careful of the timezone
# params={"from": parser.parse("2016-12-07 18:00:00 EDT").strftime('%s'),
#         "to": parser.parse("2016-12-10 00:000:00 EDT").strftime('%s'),
#         "granularity":'M1',
#         "price":'A'}
start_date = pytz.timezone("America/New_York").localize(datetime.datetime(2020, 6, 19, 6, 0, 0)).strftime("%Y-%m-%dT%H:%M:%SZ")
end_date = pytz.timezone("America/New_York").localize(datetime.datetime.now()).strftime("%Y-%m-%dT%H:%M:%SZ")
params={"from": start_date,
        "to": end_date,
        "granularity":'M1',
        "price":'A'}
# r = instruments.InstrumentsCandles(instrument="EUR_USD",params=params)
# data = client.request(r)

df_full=pd.DataFrame()
for r in InstrumentsCandlesFactory(instrument="EUR_USD",params=params):
    client.request(r)
    dat = []
    api_data=r.response.get('candles')
    if(api_data):
        for oo in r.response.get('candles'):
            print(oo)
            dat.append([oo['time'], oo['volume'], oo['ask']['o'], oo['ask']['h'], oo['ask']['l'], oo['ask']['c']])

        df = pd.DataFrame(dat)
        df.columns = ['time', 'volume', 'open', 'high', 'low', 'close']
        df = df.set_index('time')
        if df_full.empty:
            df_full=df
        else:
            df_full=df_full.append(df)
df_full.index=pd.to_datetime(df_full.index)    
# return df_full

# results= [{"time":x['time'],"closeAsk":float(x['ask']['c'])} for x in data['candles']]
# df = pd.DataFrame(results).set_index('time')

# df.index = pd.DatetimeIndex(df.index)

# df
```
